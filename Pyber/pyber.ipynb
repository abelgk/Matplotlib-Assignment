```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(ride_data, city_data, on='city')

# Display the data table for preview
combined_data.head()
```

```python
# Obtain the x and y coordinates for each of the three city types

urban_type = combined_data.loc[combined_data['type'] =='Urban']
rural_type = combined_data.loc[combined_data['type'] =='Rural',:]
suburban_type = combined_data.loc[combined_data['type'] =='Suburban',:]
total_urban_rides = urban_type.groupby(['city']).count()
total_rural_rides = rural_type.groupby(['city']).count()
total_suburban_rides = suburban_type.groupby(['city']).count()
avg_fare_urban = urban_type.groupby('city').mean()
avg_fare_rural = rural_type.groupby('city').mean()
avg_fare_sububrban = suburban_type.groupby('city').mean()
#calculate the avg fare per city

#calculate total drivers
urban_drivers = urban_type.groupby('city').sum()
rural_drivers = rural_type.groupby('city').sum()
suburban_drivers = suburban_type.groupby('city').sum()
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Build the scatter plots for each city types
urban_plot = plt.scatter(total_urban_rides['ride_id'],avg_fare_urban['fare'], 
            marker = 'o', color='lightcoral',s = urban_drivers['driver_count'].astype(int), 
            edgecolors='black',alpha=0.8, label='Urban', linewidths=1)
suburban_plot = plt.scatter(total_suburban_rides['ride_id'],avg_fare_subabrban['fare'], 
            marker = 'o', color='lightskyblue',s = suburban_drivers['driver_count'].astype(int), 
            edgecolors='black',alpha=0.8, label='Suburban', linewidths=1)
rural_plot = plt.scatter(total_rural_rides['ride_id'],avg_fare_rural['fare'], 
            marker = 'o', color='gold',s = rural_drivers['driver_count'].astype(int),
            edgecolors='black',alpha=0.8, label='Rural', linewidths=1)



# Create a legend
#plt.legend(handles=[urban_plot, suburban_plot,rural_plot], loc="best")
plt.legend(scatterpoints=1, frameon=True, labelspacing=1, title='City Types')
plt.tight_layout()

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City )")
plt.grid(color='gray', linestyle='-', linewidth=0.5)
 
# Incorporate a text label regarding circle size

# Save Figure
plt.savefig("Images/RideSharingData.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fare = combined_data['fare'].sum()
urban_percent = (urban_type['fare'].sum()/total_fare)*100
#print(f"{urban_percent:.0f}%")
rural_percent = (rural_type['fare'].sum()/total_fare)*100
#print(f"{rural_percent:.0f}%")
suburban_percent = (suburban_type['fare'].sum()/total_fare)*100
#print(f"{subarban_percent:.0f}%")

# Labels for the sections of our pie chart
labels = ["Rural", "Suburban","Urban"]

# The values of each section of the pie chart
sizes = [rural_percent, suburban_percent, urban_percent]

# The colors of each section of the pie chart
colors = ["gold","lightskyblue","lightcoral"]

# to seperate the Pie section from the others
explode = (0, 0, 0.1)
# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("Images/FaresbyCityType.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_rides = combined_data['ride_id'].count()
urban_ride_pct = (urban_type['ride_id'].count()/total_rides)*100
#print(f"{urban_ride_pct:.0f}%")
rural_ride_pct = (rural_type['ride_id'].count()/total_rides)*100
#print(f"{rural_ride_pct:.0f}%")
suburban_ride_pct = (suburban_type['ride_id'].count()/total_rides)*100
#print(f"{suburban_ride_pct:.0f}%")
# Labels for the sections of our pie chart
labels = ["Rural", "Suburban","Urban"]

# The values of each section of the pie chart
sizes = [rural_ride_pct, suburban_ride_pct, urban_ride_pct]

# The colors of each section of the pie chart
colors = ["gold","lightskyblue","lightcoral"]

# to seperate the Pie section from the others
explode = (0, 0, 0.1)
# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("Images/RidesbyCityType.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
total_drivers = city_data['driver_count'].sum()
#print(total_drivers)

grpby_type = city_data.groupby(["type"]).sum()
rural_driver_pct = (grpby_type.loc[('Rural')]/total_drivers)*100
suburban_driver_pct = (grpby_type.loc[('Suburban')]/total_drivers)*100
urban_driver_pct = (grpby_type.loc[('Urban')]/total_drivers)*100


# Labels for the sections of our pie chart
labels = ["Rural", "Suburban","Urban"]

# The values of each section of the pie chart
sizes = [rural_driver_pct, suburban_driver_pct, urban_driver_pct]

# The colors of each section of the pie chart
colors = ["gold","lightskyblue","coral"]

# to seperate the Pie section from the others
explode = (0, 0, 0.1)
# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("Images/DriversbyCityType.png")
```
